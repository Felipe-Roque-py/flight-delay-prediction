```python
import pandas as pd
```

```python


df = pd.read_csv(
    "https://siros.anac.gov.br/siros/registros/diversos/vra/2023/VRA_2023_01.csv",
    sep=';', 
    encoding='ISO-8859-1',
    on_bad_lines='skip', nrows = 10000
)
```

```python
df.head(3)
```

```python
df.rename(columns={
    'Sigla ICAO Empresa AÃ©rea': 'Sigla ICAO Empresa Aérea',
    'Empresa AÃ©rea': 'Empresa Aérea',
    'NÃºmero Voo': 'Número Voo',
    'CÃ³digo DI': 'Código DI',
    'CÃ³digo Tipo Linha': 'Código Tipo Linha',
    'Modelo Equipamento': 'Modelo Equipamento',
    'NÃºmero de Assentos': 'Número de Assentos',
    'Sigla ICAO Aeroporto Origem': 'Sigla ICAO Aeroporto Origem',
    'DescriÃ§Ã£o Aeroporto Origem': 'Descrição Aeroporto Origem',
    'Partida Prevista': 'Partida Prevista',
    'Partida Real': 'Partida Real',
    'Sigla ICAO Aeroporto Destino': 'Sigla ICAO Aeroporto Destino',
    'DescriÃ§Ã£o Aeroporto Destino': 'Descrição Aeroporto Destino',
    'Chegada Prevista': 'Chegada Prevista',
    'Chegada Real': 'Chegada Real',
    'SituaÃ§Ã£o Voo': 'Situação Voo',
    'Justificativa': 'Justificativa',
    'ReferÃªncia': 'Referência',
    'SituaÃ§Ã£o Partida': 'Situação Partida',
    'SituaÃ§Ã£o Chegada': 'Situação Chegada'
}, inplace=True)
```

```python
df.head(3)
```

```python
df = df.dropna(subset=["Partida Prevista", "Partida Real", "Chegada Prevista", "Chegada Real"])

df.isna().sum()
```

```python
df["Partida Prevista"] = pd.to_datetime(df["Partida Prevista"], format="%d/%m/%Y %H:%M")
df["Partida Real"] = pd.to_datetime(df["Partida Real"], format="%d/%m/%Y %H:%M")
df["Chegada Prevista"] = pd.to_datetime(df["Chegada Prevista"], format="%d/%m/%Y %H:%M")
df["Chegada Real"] = pd.to_datetime(df["Chegada Real"], format="%d/%m/%Y %H:%M")
```

```python
df.head(3)
```

```python
df.describe()
```

```python
# Extrair Extraindo componentes das datas

df["chegada_mes"] = df["Chegada Real"].dt.month
df["chegada_dia"] = df["Chegada Real"].dt.day
df["chegada_hora"] = df["Chegada Real"].dt.hour
df["chegada_dia_semana"] = df["Chegada Real"].dt.dayofweek  # segunda=0, domingo=6

df["chegada_dia_semana_nome"] = df["Chegada Real"].dt.day_name(locale='pt_BR')

```

```python
df.head(3)
```

```python
df.loc[
    (df["Situação Partida"] == "Atraso 30-60") |
    (df["Situação Partida"] == "Atraso 60-120") |
    (df["Situação Partida"] == "Atraso 120-240") |
    (df["Situação Partida"] == "Atraso > 240"),
    "Status"
] = "Atrasado"

df.loc[
    df["Situação Partida"] == "Antecipado",
    "Status"
] = "Antecipado"


df.loc[
    df["Situação Partida"] == "Pontual",
    "Status"
] = "Pontual"
```
